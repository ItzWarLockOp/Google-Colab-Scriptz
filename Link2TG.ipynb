```
#@title <center> <b><h2><font color='black'> â˜… Install Required Libraries  â˜… </b></h5></center>

import subprocess
import time
from IPython.display import clear_output

def check_modules_satisfied(module_names):
    for module in module_names:
        try:
            __import__(module)
        except ImportError:
            return False
    return True

def animate_characters(text, interval, num_cycles):
    characters = ["â—‰-â—Ž", "â—Ž-â—‰"]
    for _ in range(num_cycles):
        for char in characters:
            clear_output(wait=True)
            print(text + " " + char, end="")
            time.sleep(interval)

def install_dependencies(module_names):
    if check_modules_satisfied(module_names):
        print("Libraries Already Installed!")
        return

    installed_count = 0
    for module in module_names:
        process = subprocess.Popen(
            f"pip install {module}",
            shell=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE
        )
        animate_characters(f"Installing {module} :", 0.5, 5)
        process.communicate()

        if process.returncode == 0:
            clear_output(wait=True)

            installed_count += 1

    if installed_count == len(module_names):
        print("Libraries Installed Successfully!")

# Modules to install
modules = ['pyrogram', 'tgcrypto', 'nest_asyncio']

# Call the function to install the modules
install_dependencies(modules)
```

```
import re
import requests
import urllib.parse
import mimetypes
import asyncio
import aiohttp
import nest_asyncio
from pyrogram import Client
import os

def get_filename(url):
    return os.path.basename(url)

# Function to upload the file asynchronously
async def upload_file(upload_type, url, thumbnail_path=None, caption=None, api_id=None, api_hash=None, bot_token=None, chat_id=None):
    try:
        # Create a Pyrogram client
        async with Client("my_account", api_id=api_id, api_hash=api_hash, bot_token=bot_token) as app:
            # Retrieve the filename using the get_filename function
            filename = get_filename(url)

            print("Starting Engine: Pyrogramâœ¨")  # Print the engine message at the beginning
            print("")

            # Delay to simulate processing
            await asyncio.sleep(2)

            # Define the downloaded filename
            downloaded_filename = None

            # Download the file
            async with aiohttp.ClientSession() as session:
                async with session.get(url) as response:
                    if response.status == 200:
                        # Save the downloaded file using the extracted filename
                        downloaded_filename = get_filename(url)
                        print(f"Name: {downloaded_filename}")
                        with open(downloaded_filename, "wb") as f:
                            while True:
                                chunk = await response.content.read(1024 * 8)  # 8 KB chunks
                                if not chunk:
                                    break
                                f.write(chunk)

            # Check if downloaded_filename is valid before proceeding
            if downloaded_filename:

                # Define a custom progress function for sending
                async def send_progress(current, total, bar_length=10):
                    nonlocal last_progress
                    percent_complete = (current * 100) // total
                    progress_bar = "[" + "â– " * (percent_complete * bar_length // 100) + \
                                   "â–¡" * ((100 - percent_complete) * bar_length // 100) + "]"

                    # Use carriage return to overwrite the progress bar on the same line
                    print(f"\rProgress: {percent_complete}% {progress_bar}", end="", flush=True)

                last_progress = -1
                # Choose the appropriate method to send the file with the custom progress function
                if upload_type == "document":
                    # If thumbnail_path is not empty, send the thumbnail along with the document
                    if thumbnail_path:
                        await app.send_document(chat_id, downloaded_filename, caption=caption, thumb=thumbnail_path, progress=send_progress)
                    else:
                        await app.send_document(chat_id, downloaded_filename, caption=caption, progress=send_progress)
                elif upload_type == "video":
                    # If thumbnail_path is not empty, send the thumbnail along with the video
                    if thumbnail_path:
                        await app.send_video(chat_id, downloaded_filename, caption=caption, thumb=thumbnail_path, progress=send_progress)
                    else:
                        await app.send_video(chat_id, downloaded_filename, caption=caption, progress=send_progress)
                else:
                    print("Invalid upload type specified")

                # Print the final message on a new line
                print("")
                print("\nFile Uploaded SuccessfullyðŸŽ‰")
            else:
                print("Error downloading the file")
    except Exception as e:
        print("File Upload Failed: ", e)

#@markdown ----
#@markdown <center> <b><h5><font color='DarkTurquoise'> â˜… ENTER TELEGRAM CREDENTIALS â˜… </b></h5></center>

# Define a main coroutine to run the upload
async def main():
    # Fill all required options
    api_id = "" #@param {type:"string"}
    api_hash = "" #@param {type:"string"}
    bot_token = "" #@param {type:"string"}
    chat_id = "ENTER CHANNEL USERNAME" #@param {type:"string"}

#@markdown ----
#@markdown <center> <b><h5><font color='Goldenrod'> â˜… ENTER URL AND UPLOAD TYPE â˜… </b></h5></center>


    # Choose File Type and Enter Url
    url = "https://directURL.com/d=file.com" #@param {type:"string"}
    upload_type = "document" #@param ["document", "video"]

#@markdown ----
#@markdown <center> <b><h5><font color='magenta'> â˜… Aditional Settings (optional) â˜… </b></h5></center>
    thumbnail_path = "" #@param {type:"string"}
    caption = "" #@param {type:"string"}

    # Call the upload_file function
    await upload_file(upload_type, url, thumbnail_path=thumbnail_path, caption=caption, api_id=api_id, api_hash=api_hash, bot_token=bot_token, chat_id=chat_id)

# Allow running asyncio code within the existing Colab event loop
nest_asyncio.apply()

# Run the main coroutine within the Colab event loop
asyncio.get_event_loop().run_until_complete(main())
```
